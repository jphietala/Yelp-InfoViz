# Explorative data analysis of cleaned data frame

This notebook is meant for EDA of the dataframe we ended up with after cleaning the data from irrelevant attributes, missing values, duplicates and outliers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('..\Yelp Dataset\clean_final.csv')
df.head()
```

```python
df.info()
```

```python
# Setting float format options to show floats with three desimals
pd.set_option('display.float_format', lambda x: '%.3f' % x)

df.describe(include='all')
```

Let's start data exploration by subsetting the original dataframe to subsets of numerical, categorical and coordinate data.

```python
# Lists containing attribute labels

numerical = ['review_stars', 'restaurant_stars', 'user_review_count', 'fans', 'average_stars', 'restaurant_review_count', 'cat_avg_rating', 'state_avg_rating', 'state_cat_avg_rating']

categorical = ['city', 'state', 'postal_code', 'category']

coordinates = ['latitude', 'longitude']
```

Now we can subset the dataframe to these different types by filtering by column labels

```python
print(df.shape)

df_numerical = df[numerical]
print(df_numerical.shape)

df_categorical = df[categorical]
print(df_categorical.shape)

df_coordinates = df[coordinates]
print(df_coordinates.shape)
```

```python
# Setting up seaborn for plotting
sns.set(palette="deep", font_scale=1.1)
```

### Numerical variables

```python
# Plotting distributions of numerical variables
df_numerical.hist(bins=50, figsize=(15, 20), layout=(3, 3))
print('Values are on the x-axis and counts are on the y-axis')
```

### Categorical variables

```python
# Review counts of different states

plt.figure(figsize=(14,8))
chart = sns.countplot(data=df_categorical, x='state', order = df_categorical['state'].value_counts().index)
chart.set_xticklabels(labels=df_categorical['state'].value_counts().index, rotation=90)

df_categorical.state.value_counts()
```

```python
# Review counts of different cities
df_categorical.city.value_counts()
```

```python
# Review counts of different postal codes
df_categorical.postal_code.value_counts()
```

```python
# Review counts of different cuisiness

plt.figure(figsize=(14,8))
chart = sns.countplot(data=df_categorical, x='category', order = df_categorical['category'].value_counts().index)
chart.set_xticklabels(labels=df_categorical['category'].value_counts().index, rotation=90)

df_categorical.category.value_counts()
```

### Coordinates

```python
sns.jointplot(x='longitude', y='latitude', data=df_coordinates, height=12)
```
