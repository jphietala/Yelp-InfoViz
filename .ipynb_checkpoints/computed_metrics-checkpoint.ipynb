# Computing metrics

In this notebook metrics for needed attributes are computed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('..\Yelp Dataset\clean_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Category aggregated metrics, computing mean for each restaurant category

vc = df['category'].unique() # All unique restaurant categories

for i in vc:
    avg_rating = np.mean(df.loc[df['category'] == i]['restaurant_stars'])
    df.loc[df['category'] == i,'cat_avg_rating'] = avg_rating

df.head()
```

```python
#State aggregated metrics, computing mean for each state

df['state_avg_rating'] = np.nan
vc = df['state'].unique()

for i in vc:
    avg_rating = np.mean(df.loc[df['state'] == i]['restaurant_stars'])
    df.loc[df['state'] == i,'state_avg_rating'] = avg_rating

df.head()
```

```python
#State and category aggregated metric, computing mean for each state and category combination

df['state_cat_avg_rating'] = np.nan
vc = df['state'].unique()

for i in vc:
    for j in df.loc[df['state'] == i]['category'].unique():
        avg_rating = np.mean(df.loc[(df['state'] == i) & (df['category'] == j)]['restaurant_stars'])
        df.loc[(df['state'] == i) & (df['category'] == j),'state_cat_avg_rating'] = avg_rating
df.head()
```
