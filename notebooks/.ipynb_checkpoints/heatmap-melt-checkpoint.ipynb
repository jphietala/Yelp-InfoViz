```python
import pandas as pd
```

```python
df = pd.read_csv('../visualization/data/heatmap.csv', index_col=False)
df.head()
```

```python
df = df.melt(id_vars=['States'], var_name='Cuisine', value_name='Rating')
print(df)
```

```python
df['Rating'] = df.Rating.round(2)
df.head()
```

```python
df.to_csv('../visualization/data/heatmap_melt.csv', index=False)
```
