```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("clean.csv")
```

```python
dataset.head()
```

```python
# Recalculate ratings
vc = dataset['business_id'].unique()
dataset['recal_stars'] = np.nan
print(len(vc))
for num,i in enumerate(vc):
    data = dataset.loc[dataset['business_id'] == i]['review_stars']
    dataset.loc[dataset['business_id'] == i,'recal_stars'] = np.mean(data)
    if num % 50 == 0:
        print(num/len(vc)*100, "%")
        
    
```

```python
dataset.to_csv('clean_avgs.csv',index=None)

dataset.head(20)
```

```python
dataset['restaurant_stars'] = dataset['recal_stars']
```

```python
rest = dataset[['business_id','state','restaurant_review_count', 'category', 'restaurant_stars']].drop_duplicates()
revs = dataset[['user_id', 'user_review_count', 'user_average_stars']].drop_duplicates()
```

```python
# Category
print(rest.shape)
vc = rest['category'].unique()
num_cat = len(vc)
cats = list(vc)
for i in vc:
    data = rest.loc[rest['category'] == i]['restaurant_stars']
    rest.loc[rest['category'] == i,'cat_median'] = np.median(data)
    rest.loc[rest['category'] == i,'cat_25'] = np.percentile(data,25)
    rest.loc[rest['category'] == i,'cat_75'] = np.percentile(data,75)
    rest.loc[rest['category'] == i,'cat_max'] = np.max(data)
    rest.loc[rest['category'] == i,'cat_min'] = np.min(data)
    #print(i,rest.loc[rest['category'] == i]['cat_median'])
```

```python
#State
print(rest.shape)
#rest['state_rating'] = np.nan
vc = rest['state'].unique()
num_states = len(vc)
for i in vc:
    data = rest.loc[rest['state'] == i]['restaurant_stars']
    rest.loc[rest['state'] == i,'state_median'] = np.median(data)
    rest.loc[rest['state'] == i,'state_25'] = np.percentile(data,25)
    rest.loc[rest['state'] == i,'state_75'] = np.percentile(data,75)
    rest.loc[rest['state'] == i,'state_max'] = np.max(data)
    rest.loc[rest['state'] == i,'state_min'] = np.min(data)
    #print(i,rest.loc[rest['state'] == i]['state_median'])
```

```python
#State and category
print(rest.shape)
#rest['state_cat_rating'] = np.nan
vc = rest['state'].unique()
heatmap = np.zeros((num_states,num_cat))
for num,i in enumerate(vc):
    cuisines = rest.loc[rest['state'] == i]['category'].unique()
    
    for j in cuisines:
        #print(j)
        data = rest.loc[(rest['state'] == i) & (rest['category'] == j)]['restaurant_stars']
        heatmap[num,cats.index(j)] = np.mean(data)
        rest.loc[(rest['state'] == i) & (rest['category'] == j),'state_cat_median'] = np.median(data)
        rest.loc[(rest['state'] == i) & (rest['category'] == j),'state_cat_25'] = np.percentile(data,25)
        rest.loc[(rest['state'] == i) & (rest['category'] == j),'state_cat_75'] = np.percentile(data,75)
        rest.loc[(rest['state'] == i) & (rest['category'] == j),'state_cat_max'] = np.max(data)
        rest.loc[(rest['state'] == i) & (rest['category'] == j),'state_cat_min'] = np.min(data)
        
        #print(i,j)
```

```python
import matplotlib.pyplot as plt
stats = []
vc = rest['state'].unique()
print(vc.shape)
for i in vc:
    data = rest.loc[rest['state'] == i].iloc[0]
    item = {}

    item["label"] = i # not required
    #item["mean"] = 5 # not required
    item["med"] = data['state_median']
    item["q1"] = data['state_25']
    item["q3"] = data['state_75']
    #item["cilo"] = 5.3 # not required
    #item["cihi"] = 5.7 # not required
    item["whislo"] = data['state_min']
    item["whishi"] = data['state_max']
    #item["fliers"] = [] # required if showfliers=True
    #print(item)
    stats.append(item)

fig, axes = plt.subplots(1, 1)
axes.bxp(stats,showfliers=False)
axes.set_title('Review distribution')
y_axis = [ 1, 2, 3, 4, 5]
y_values = [ "1", "2", "3", "4", "5"]
plt.yticks(y_axis, y_values)
plt.show()
```

```python
year = np.zeros(366,dtype=int)
```

```python
dates = [dat[5:] for dat in date['review_date']]
```

```python
from collections import Counter
dates_counts = Counter(dates)
```

```python
import datetime
del dates_counts['02-29']
for date in dates_counts:
    dt = datetime.datetime(year=2019,month=int(date[:2]), day=int(date[3:]))
    #print(dt.timetuple().tm_yday)
    year[dt.timetuple().tm_yday] = dates_counts[date]
year = year[1:]
```

```python
pd_year = pd.DataFrame({'revs_aday': year})
```

```python
pd_year.to_csv('rev_aday.csv')
```
